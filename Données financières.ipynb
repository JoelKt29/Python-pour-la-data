```python
import yfinance as yf
import pandas as pd
entreprises = ["MSFT", "NVDA",  "AAPL",
    "AMZN", "GOOGL", "GOOG", "META", "TSLA", "BRK-B", "AVGO",
    "JPM", "LLY", "V", "MA", "UNH", "XOM", "WMT", "COST", "HD","NFLX"]

data = yf.download(
    tickers = entreprises,
    start = "2024-01-02", #on commence le 2 car le 1er janvier est férié
    end = "2025-01-01",
    auto_adjust = True,     #ajustement du prix des actions (une histoire de dividende, à creuser)
    interval = "1d",
    group_by="ticker")

donnees_financieres_df = data.stack(level=0, future_stack=True).reset_index() #réorganisation du dataframe pour être lisible
print(donnees_financieres_df.head())
```
